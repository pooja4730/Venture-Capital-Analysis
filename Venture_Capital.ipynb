# **Venture Capital**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset
df = pd.read_csv('investments_VC.csv', encoding='latin1')
```

```python
# Clean column names
df.columns = df.columns.str.strip()
```

```python
# Clean funding column (remove commas and convert to number)
df['funding_total_usd'] = (
    df['funding_total_usd']
    .astype(str)
    .str.replace(',', '')
    .str.replace('-', '')
)

df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'], errors='coerce')
```

```python
#convert date
df['first_funding_at'] = pd.to_datetime(df['first_funding_at'], errors='coerce')
```

```python
#extract year
df['funding_year'] = df['first_funding_at'].dt.year
```

```python
#dropping rows where funding is missing
df_clean = df.dropna(subset=['funding_total_usd', 'funding_year'])

print("Rows after proper cleaning:", df_clean.shape)
```

Funding Trends Over Years

```python
funding_trend = df_clean.groupby('funding_year')['funding_total_usd'].sum()

plt.figure()
funding_trend.plot()
plt.title("Total VC Funding Over Years")
plt.xlabel("Year")
plt.ylabel("Funding (USD)")
plt.show()
```

Top 10 Countries by Funding

```python
top_countries = df_clean.groupby('country_code')['funding_total_usd'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_countries.plot(kind='bar')
plt.title("Top 10 Countries by VC Funding")
plt.xlabel("Country")
plt.ylabel("Funding (USD)")
plt.show()
```

Top 10 Markets (Sectors)

```python
top_markets = df_clean.groupby('market')['funding_total_usd'].sum().sort_values(ascending=False).head(10)

plt.figure()
top_markets.plot(kind='bar')
plt.title("Top 10 Markets by VC Funding")
plt.xlabel("Market")
plt.ylabel("Funding (USD)")
plt.show()
```

Number of Startups Funded Each Year

```python
startups_year = df_clean.groupby('funding_year')['name'].count()

plt.figure()
startups_year.plot()
plt.title("Number of Startups Funded Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Startups")
plt.show()
```

```python
import seaborn as sns

fig, axes = plt.subplots(3, 2, figsize=(18, 15))
fig.suptitle('Venture Capital Investment Behavior and Trends')

# 1. Distribution of total funding
sns.histplot(ax=axes[0, 0], x='funding_total_usd', data=df_clean, bins=30, kde=True, color='skyblue')
axes[0, 0].set_title('Distribution of Total Funding (USD)')

# 2. Funding trend over years
sns.histplot(ax=axes[0, 1], x='funding_year', data=df_clean, bins=20, kde=False, color='orange')
axes[0, 1].set_title('Funding Trend Over Years')

# 3. Top markets receiving funding
top_markets = df_clean['market'].value_counts().head(10)
sns.barplot(ax=axes[1, 0], x=top_markets.values, y=top_markets.index, color='green')
axes[1, 0].set_title('Top 10 Markets by Number of Startups')

# 4. Top countries receiving funding
top_countries = df_clean['country_code'].value_counts().head(10)
sns.barplot(ax=axes[1, 1], x=top_countries.values, y=top_countries.index, color='red')
axes[1, 1].set_title('Top 10 Countries by Number of Startups')

# 5. Average funding per market
avg_market_funding = df_clean.groupby('market')['funding_total_usd'].mean().sort_values(ascending=False).head(10)
sns.barplot(ax=axes[2, 0], x=avg_market_funding.values, y=avg_market_funding.index, color='purple')
axes[2, 0].set_title('Average Funding per Market')

# 6. Number of startups funded per year
startups_per_year = df_clean['funding_year'].value_counts().sort_index()
sns.lineplot(ax=axes[2, 1], x=startups_per_year.index, y=startups_per_year.values, color='brown')
axes[2, 1].set_title('Number of Startups Funded Per Year')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
